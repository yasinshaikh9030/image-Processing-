```python

negative_img = np.zeros(img.shape, dtype=np.uint8)

for i in range(img.shape[0]):        # height
    for j in range(img.shape[1]):    # width
        negative_img[i, j] = 255 - img[i, j]
```

Contrast stretching improves image contrast by expanding pixel intensity values to the full range of 0 to 255.

```python
#contrast streching
imgGray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray,cmap='gray')
plt.axis("off")


```

```python
min_val=np.min(imgGray)
max_val=np.max(imgGray)
print(min_val,max_val)
```

```python
stretched_img=(imgGray-min_val)/(max_val-min_val)*255
stretched_img = stretched_img.astype(np.uint8)
plt.imshow(streched_img,cmap='gray')
plt.axis("off")
```
